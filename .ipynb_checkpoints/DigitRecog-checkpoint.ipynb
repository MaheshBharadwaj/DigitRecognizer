```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('./Data/train.csv')
```

```python
dataset.info()
```

```python
dataset
```

```python
data = dataset[['pixel'+str(i) for i in range(784)]]
```

```python
data
```

```python
d  = data.iloc[:,:].values
d
```

```python
d_biased = np.c_[np.ones(len(d)),d]
```

```python

d_biased
```

```python
truth = dataset['label']
```

```python
truth
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import metrics
from sklearn.model_selection import train_test_split
from keras.utils import np_utils
```

```python
model = Sequential()
model.add(Dense(units = 785, input_dim = 785, activation = 'relu'))
model.add(Dense( units = 785, activation = 'relu'))
model.add(Dense( units = 785, activation = 'relu'))
model.add(Dense( units = 785, activation = 'relu'))
model.add(Dense( units = 785, activation = 'relu'))
model.add(Dense( units = 785, activation = 'relu'))
model.add(Dense( units = 785, activation = 'relu'))
model.add(Dense( units = 785, activation = 'relu'))
model.add(Dense( units = 785, activation = 'relu'))
model.add(Dense( units = 785, activation = 'relu'))
model.add(Dense( units = 10,activation = 'sigmoid'))
model.compile(loss='binary_crossentropy', optimizer=Adam(lr = 0.001), metrics=['accuracy'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(d_biased, truth, test_size=0.20, random_state=42)
y_train = np_utils.to_categorical(y_train,10)
y_test = np_utils.to_categorical(y_test,10)
```

```python
m = model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs = 4,batch_size = 32,verbose = 2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(m.history['acc'])
plt.plot(m.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Iteration')
plt.legend(['Train M', 'Test M'], loc='upper left')
plt.show()
```

```python
model.save('./Model1.h5')
```
